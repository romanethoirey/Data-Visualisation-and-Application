```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from shapely.geometry import Point, Polygon
import geopandas as gpd
```

```python
data = pd.read_csv("Meteorite_Landings.csv")

data.head()
```

```python
data = data.groupby('nametype').get_group('Valid').copy()
data.dropna(inplace=True)
data = data.reset_index()
data = data.drop(columns=["index"])
data = data.drop(columns=["GeoLocation"])
data.head(10)
```

```python
data.dtypes
```

```python
# merge coordinate and convert into points
data['Coordinates'] = list(zip(data.reclong, data.reclat))
data['Coordinates'] = data['Coordinates'].apply(Point)

#convert into geapanda Dataframe
crs = {'init': 'epsg:4326'}
data = gpd.GeoDataFrame(data, crs = crs, geometry='Coordinates')
data.head(15)
```

```python
# new column to store iso_a3 values
data['continent'] = 'none'
data['iso_a3'] = 'default value'
```

```python
data = pd.read_csv("data_meteori.csv")

data.year = pd.to_datetime(data.year, errors = 'coerce')
data['time'] = data.year
data.set_index('time',inplace=True)

# merge coordinate and convert into points
data['Coordinates'] = list(zip(data.reclong, data.reclat))
data['Coordinates'] = data['Coordinates'].apply(Point)

#convert into geapanda Dataframe
crs = {'init': 'epsg:4326'}
data = gpd.GeoDataFrame(data, crs = crs, geometry='Coordinates')
data.head()
```
